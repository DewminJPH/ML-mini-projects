```python
# This dataset is highly unbalanced.
# 0 --> normal transaction
# 1 --> fraudulent transaction
```

```python
# seperating the data for analysis
legit = credit_card_data[credit_card_data.Class==0]
fraud = credit_card_data[credit_card_data.Class==1]
```

# Dealing with unbalanced data
# Under-sampling

```python
# Build a sample dataset containing similar distribution of normal transactions and fraudulent transactions
```

```python
# Number of Fraudulent Transactions -> 492
```

```python
legit_sample = legit.sample(n=492)
```

```python
# Conacatenating the two DataFrames
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0) # axis = 0 : add the column with rowwise
```

```python
new_dataset.head()
```

```python
new_dataset.shape
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby("Class").mean()
```

# Splitting the data into Features & Targets

```python
x = new_dataset.drop(columns="Class", axis=1)
y = new_dataset["Class"]
```

```python
print(x)
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, stratify=y, random_state=2)
```

# Model Training

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
# model evaluation
train_prediction = model.predict(x_train)
train_data_accuracy = accuracy_score(y_train, train_prediction)
print("Training data accuracy :", train_data_accuracy)
```

```python
test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(y_test, test_prediction)
print("Testing data accuracy :", test_data_accuracy)
```
