# Handling the missing values(data imputation method)
- Mean --> average value
- Mode --> Most Repeated value

```python
big_mart_data['Item_Weight'].mean()
```

```python
# filling the missing values in "Item_Weight" column with "Mean" value
big_mart_data['Item_Weight'].fillna(big_mart_data['Item_Weight'].mean(), inplace=True)
```

```python
big_mart_data['Outlet_Size'].mode()
```

```python
big_mart_data.isnull().sum()
```

```python
# Replacing the missing values in "Outlet_Size" with mode
mode_of_Outlet_Size= big_mart_data.pivot_table(values='Outlet_Size', columns="Outlet_Type", aggfunc=(lambda x: x.mode()[0]))
```

```python
print(mode_of_Outlet_Size)
```

```python
missing_values = big_mart_data['Outlet_Size'].isnull()
```

```python
print(missing_values) # true- value is null ,, false - value is there
```

```python
# Replace the missing values with mode
big_mart_data.loc[missing_values,'Outlet_Size'] = big_mart_data.loc[missing_values, 'Outlet_Type'].apply(lambda x: mode_of_Outlet_Size[x])
```

# Data Preprocessing

```python
big_mart_data.head()
```

```python
big_mart_data['Item_Fat_Content'].value_counts()
```

```python
# here we can see same the values are exist with different names. So we need to merge them with suitable values
big_mart_data.replace({'Item_Fat_Content':{'low fat':'Low Fat','LF':'Low Fat', 'reg':'Regular'}},inplace=True)
```

```python
big_mart_data['Item_Fat_Content'].value_counts()
```

* Label Encoding

```python
encoder = LabelEncoder()
```

```python
big_mart_data['Item_Identifier']= encoder.fit_transform(big_mart_data['Item_Identifier'])

big_mart_data['Item_Fat_Content']= encoder.fit_transform(big_mart_data['Item_Fat_Content'])

big_mart_data['Item_Type']= encoder.fit_transform(big_mart_data['Item_Type'])

big_mart_data['Outlet_Identifier']= encoder.fit_transform(big_mart_data['Outlet_Identifier'])

big_mart_data['Outlet_Size']= encoder.fit_transform(big_mart_data['Outlet_Size'])

big_mart_data['Outlet_Location_Type']= encoder.fit_transform(big_mart_data['Outlet_Location_Type'])

big_mart_data['Outlet_Type']= encoder.fit_transform(big_mart_data['Outlet_Type'])
```

```python
big_mart_data.head()
```

# Splitting features and target

```python
x = big_mart_data.drop(columns='Item_Outlet_Sales', axis=1)
y = big_mart_data['Item_Outlet_Sales']
```

```python
print(x)
print(y)
```

```python
x_train,x_test,y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

# Machine Learning Model Training
- XGBoost Regressor

```python
model = XGBRegressor()
```

```python
model.fit(x_train,y_train)
```

# Evaluation

```python
# prediction on training data
training_data_prediction = model.predict(x_train) 
```

```python
# R squared value
r2_train = metrics.r2_score(y_train, training_data_prediction)
print('R sqaured value for trained data:',r2_train)
```

```python
testing_data_prediction = model.predict(x_test)
r2_test = metrics.r2_score(y_test, testing_data_prediction)
print('R sqaured value for test data:',r2_train)
```
